```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
msudata = pd.read_excel('PS3_data.xlsx', header=0, index_col=None)
msudata.head(n=10)
```

```python
msudata[["college", "year"]]
```

```python
msuyears = msudata.loc[0:18, 'year']
msuapps = msudata.loc[0:18, 'total_apps']
synthyrs = msudata.loc[38:56, 'year']
synthapps = msudata.loc[38:56, 'total_apps']

plt.style.use('default')
fig, ax = plt.subplots()
plt.plot(msuyears, msuapps, axes = ax, label="Michigan State University", c='g')
plt.plot(synthyrs, synthapps, axes = ax, label="Synthetic Michigan State", c='b', ls='--')
plt.legend()
ax.set_xlim([2001, 2019])
ax.set(title='Undergraduate Applications', xlabel='Year', ylabel='Number of Applications')
ax.axvline(x=2017, c='k')

fig.savefig('msu_apps.png', transparent=False, dpi=800, bbox_inches="tight")
```

```python
msuread75 = msudata.loc[0:18, 'SAT_read75']
psuread75 = msudata.loc[19:37, 'SAT_read75']
```

```python
n = 19
r = np.arange(n)
width = 0.25

plt.style.use('seaborn-darkgrid')
fig, ax = plt.subplots()
plt.bar(r, msuread75, width=width, color='C2', label = "Michigan State")
plt.bar(r + width, psuread75, width=width, color='b', label = "Penn State")

plt.xlabel("Year")
plt.ylabel("SAT Reading Score")
plt.title("SAT Reading Scores: 75th Percentile")

plt.xticks(r + width/2,['2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008',
                       '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016',
                       '2017', '2018', '2019'])
plt.ylim([0, 850])
plt.legend()
fig.autofmt_xdate(rotation=90, ha='center')
plt.savefig('msu_psu_read75pct.png')
plt.show()
```
