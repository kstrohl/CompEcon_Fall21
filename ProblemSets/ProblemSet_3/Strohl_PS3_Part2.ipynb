```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
msudata2 = pd.read_excel('nassar_full.xlsx', header=0, index_col=None)
msudata2.head(n=10)
```

```python
applications  = msudata2.sort_values(by=["year", "total_apps", "college"])[["year", "total_apps", "college"]]
applications
```

```python
table = applications[-10:]
table
```

```python
y = table.total_apps
x = table.college

plt.style.use('seaborn-whitegrid')
fig, ax = plt.subplots()
plt.barh(x, y)
plt.xlabel("Number of Applications")
plt.title("Colleges with Highest Number of Applications in 2019")
plt.savefig('college_apps2.png', transparent=False, dpi=800, bbox_inches="tight")
plt.show()
```
