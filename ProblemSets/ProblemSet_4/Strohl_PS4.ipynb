### MLE Estimation

```python
def MLE(params):
    '''
    This function replicates the model specified with the exception of the 'hispanic' indicator variable.
    
    Args: 
        params (tuple): model parameters
        
    Returns: 
        ln_wage (scalar): The natural log of hourly wages given the parameters.
    '''
    cons, beta1, beta2, beta3, beta4, beta5 = params
    
    coeff = [cons, beta1, beta2, beta3, beta4, beta5]
    
    lnwage = np.dot(x, coeff)
    neg_LL = -np.sum(stats.norm.logpdf(y, lnwage))
    return(neg_LL)
```
